```python
from matplotlib import pyplot as plt
import cv2
```

```python
img = cv2. imread(r"/content/CH1.jpg")
```

```python
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img2)
plt.axis("off")
plt.title("Original Image")
```

```python
img.shape
```

```python
# grayscale
gray= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
plt.axis("off")
plt.title("GrayScale Image")
```

```python
# Blur
blur= cv2.GaussianBlur(img,(5,5),cv2.BORDER_DEFAULT)
plt.imshow(blur)
plt.axis("off")
plt.title("GaussianBlur Image")
```

```python
# Edge cascade
canny= cv2.Canny(img,125,255)
plt.imshow(canny)
plt.axis("off")
plt.title("Canny Image")
```

```python
# Dilating the image
dilated = cv2.dilate(canny, (7,7), iterations=3)
plt.imshow(dilated)
plt.axis("off")
plt.title("Dilate Image")
```

```python
ret, thresh = cv2.threshold(gray, 125, 255, cv2.THRESH_BINARY)  
plt.imshow(thresh)
plt.axis("off")
plt.title("thresh Image")
```

```python
plt.figure(figsize=(20,20))
plt.subplot(1,5,1)
plt.imshow(img2)
plt.axis("off")
plt.title("Original Image")
plt.subplot(1,5,2)
plt.imshow(gray)
plt.axis("off")
plt.title("GrayScale Image")
plt.subplot(1,5,3)
plt.imshow(blur)
plt.axis("off")
plt.title("GaussianBlur Image")
plt.subplot(1,5,4)
plt.imshow(canny)
plt.axis("off")



plt.show()
```
